```python
#import first required libraries

import pandas as pd
import numpy as np
import warnings


#link to the required csv-file

data=pd.read_csv('fifa21_male2.csv') 

warnings.filterwarnings('ignore') #the red warning does not show up any more

data1=data.copy()

```

```python
data1.info(verbose=True)
```

```python

#drop columns which are not needed for sure

#maybe we keep POT and Growth for correlation

to_drop=['Player Photo','Club Logo','Flag Photo', 'POT','Team & Contract','Growth','Gender']

data1.drop(to_drop, inplace=True, axis=1)
```

```python
#set index player id and check if it is unique


data1['ID'].is_unique
data1=data1.set_index('ID')

data1.sort_values(by=['ID'],ascending = True)
```

```python
#standardize the columns - lower case and underscore instead of space


data1.columns = [column.lower().replace(' ', '_') for column in data1.columns]
```

```python
#look for columns with NaN
pd.options.display.min_rows = 30000

data1.isna().sum().sort_values(ascending = False)
```

```python
#replace NaN of important colums with data which makes sense

data1['position']=data1['position'].replace(np.NAN,'only_1_pos')
data1['loan_date_end']=data1['loan_date_end'].replace(np.NAN,'no_loan_player')
data1['club']=data1['club'].replace(np.NAN,'without_a_club')

```

```python
#before droppint columns usually you should check the correlation first!!!

#data1.dropna(subset=['volleys'],inplace=True)
data1.dropna(subset=['composure'],inplace=True)
```

```python
#try to clean the 'joined' column from NaT

data1.fillna(0)
```

```python
data1['joined']
```

```python
#check NaN again - special case 'joined'

data1.isna().sum().sort_values(ascending = False)
```

```python
#clean column wage


data1['wage'] = data1['wage'].str.replace("€", "").str.replace("K", "000").astype(int)
```

```python
data1['wage'].sort_values(ascending = False)
```

```python
#clean column 'value'


def value_conv(x):
    new = []
    for i in x:
        ending = i[-1]
        if ending is 'M':
            i = float(''.join(i[1:-1]))
            i *= 1000000
        elif ending is 'K':
            i = float(''.join(i[1:-1]))
            i *= 1000
        else:
            i = 0
        new.append(i)
    return new

data1['value'] = value_conv(list(data1['value']))
```

```python
#clean release clause
data1['release_clause'] = value_conv(list(data1['release_clause']))
```

```python
#clean column 'joined'


data1['joined'] = pd.to_datetime(data1['joined'], errors='coerce')
```

```python
#generate a new column 'contract_start' and add it to the dataframe

contract_start=[]
for i in data1['contract']:
    i=i[0:4]
    contract_start.append(i)
list(contract_start)
data1['contract_start']=contract_start
```

```python
#generate a new column 'contract_end' and add it to the dataframe


contract_end=[]
for i in data1['contract']:
    i=i[7:]
    contract_end.append(i)
list(contract_end)

data1['contract_end']=contract_end
```

```python
data1['contract_end']
```

```python
#generate a file for your overview reports and filters


report1=data1[['name','value','ova','age','wage','release_clause','bp','nationality','foot','joined','club','contract_start','contract_end','loan_date_end','mentality']]

report1=report1.sort_values( by=['value'],ascending=True)


report1

report1.to_csv('report1.csv')

```
