```python
#import required libraries

import pandas as pd
import numpy as np
import warnings


#link to the required csv-file

report1=pd.read_csv('report1.csv') 

warnings.filterwarnings('ignore') #the red warning does not show up any more

import seaborn as sns #visualisation
import matplotlib.pyplot as plt #visualisation
%matplotlib inline 
sns.set(color_codes=True)
```

```python
#Assumption: present to the management of a football club the most interesting active players
#as already experienced the presentation should be as flexibile as possible to give the possibility to response to
#spontaneous requests


report1=report1.set_index('ID')
report1
```
